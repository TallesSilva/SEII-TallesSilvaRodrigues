```python
def f(x):
    return x*2

x = np.linspace(0, 10, 100)
y = f(x)
dydx = np.gradient(y,x)
y_integral = np.cumsum(y) * (x[1]-x[0])
```

```python
plt.plot(x,y)
plt.plot(x,dydx)
plt.plot(x,y_integral)
```
